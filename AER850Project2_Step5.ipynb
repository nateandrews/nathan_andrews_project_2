```python
#Step 5 - Model Testing

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import os

#loading previous model
model_path = '/content/drive/MyDrive/850 Project 2 Data/model.keras'
test_model = tf.keras.models.load_model(model_path)
print("Model loaded successfully")

#defining path to test images
test_images = {
    "crack": "/content/drive/MyDrive/850 Project 2 Data/nathan_andrews_project_2-main/Data/test/crack",
    "missing_head": "/content/drive/MyDrive/850 Project 2 Data/nathan_andrews_project_2-main/Data/test/missing-head",
    "paint_off": "/content/drive/MyDrive/850 Project 2 Data/nathan_andrews_project_2-main/Data/test/paint-off"
}

#defining class names based off training
class_names = ["crack","missing-head","paint-off"]

#function to preprocess and predict on an image
def process_and_predict(image_path):
  img = load_img(image_path, target_size=(100, 100))

  #converting image to array
  img_array = img_to_array(img)/255.0
  img_array = np.expand_dims(img_array, axis = 0)

  #to make a prediction
  predictions = test_model.predict(img_array)
  predicted_class = np.argmax(predictions, axis = 1)[0]
  predicted_class_name = class_names[predicted_class]
  confidence = predictions[0][predicted_class]
  return predicted_class_name, confidence, img

#going through each test image predicting and displaying the results
for label, image_dir in test_images.items():
    for image_file in os.listdir(image_dir):
        image_path = os.path.join(image_dir, image_file)

        #checking if its a file before processing
        if os.path.isfile(image_path):
            predicted_class_name, confidence, img = process_and_predict(image_path)

    #displaying prediction and image
    plt.figure()
    plt.imshow(img)
    plt.title(f"Actual: {label.replace('_','').title()}, Predicted: {predicted_class_name.title()}({confidence*100:.2f}%)")
    plt.axis('off')
    plt.show()

    #printing results
    print(f"Actual class: {label}")
    print(f"Predicted class: {predicted_class_name}")
    print(f"Confidence: {confidence*100:.2f}%")


```
